#### AFVM1 : 
1. Compléter et tester le code suivant qui implémente une pile.

```python
def pileVide():
    """None -> pile
    renvoie une pile vide"""
    return []

def estVide(p):
    """pile -> bool
    renvoie un booléen disant si la pile est vide"""
    return len(p)==0

def empiler(p, elt):
    """pile,elt -> None"""
    p.append(elt)


def depiler(p):
    """pile -> elt
    renvoie le sommmet de la pile et modifie la pile"""
    assert len(p)>0
    s = p.pop()
    return s
    
# tests : créer une pile vide p1, empiler 2 éléments, en dépiler un, tester si la pile est vide
p1 = pileVide()
print(estVide(p1))
empiler(p1, 1)
empiler(p1,2)
print(p1)
depiler(p1)
print(estVide(p1))
```

2. Ecrire une fonction qui renvoie le sommet de la pile (sans l'enlever) et une fonction qui compte le nombre d'éléments de la pile.

```python
def sommetPile(p):
    if len(p) > 0:
        return p[-1]

def taillePile(p):
    return len(p)
    
# tests 
print(sommetPile(p1))
print(taillePile(p1))
```

2. Ecrire une fonction `renversePile` qui prend en argument une pile `p` et qui renvoie une autre pile avec les les éléments empilés dans l'ordre inverse. La pile `p`peut être vidée. 

```python
def renversePile(p):
    p1 = pileVide()
    while not estVide(p):
        s = depiler(p)
        empiler (p1,s)
    return p1
p = [1,2,3]
renversePile(p)

    
```

3. Ecrire une fonction `affichePile` qui prend en argument une pile `p` et qui renvoie une chaine de caractères représentant la pile avec les éléments les uns au-dessus des autres. Par exemple, la pile [1,2,3] sera représentée par :<br>
3 <br>2 <br>1

```python
def affichePile(p):
    resultat = ''
    while not estVide(p):
        s = depiler(p)
        resultat += str(s) + '\n'
    return resultat
    
# tests 
p = [1,2,3]
print(affichePile(p))
```

#### AFVM4 : implémentation d'une file.
1. Ecrire un code permettant d'implémenter une file.

```python
def fileVide():
    """None -> file
    renvoie une file vide"""
    return []

def estVide(f):
    """file -> bool
    renvoie un booléen disant si la pile est vide"""
    return len(f)==0

def enfiler(f, elt):
    """file,elt -> None"""
    f.append(elt)


def defiler(f):
    """file -> elt
    renvoie le sommmet de la file et modifie la file"""
    assert len(f)>0
    s = f.pop(0)
    return s
    
# tests : créer une pile vide p1, empiler 2 éléments, en dépiler un, tester si la pile est vide
f1 = fileVide()
print(estVide(f1))
enfiler(f1, 1)
enfiler(f1,2)
print(f1)
defiler(f1)
print(estVide(f1))
f1
```

2. Ecrire deux fonctions qui renvoie la queue (sans l'enlever) et qui compte le nombre d'éléments de la file.

```python
def sommetfile(f):
    if len(f) > 0:
        return f[0]

def tailleFile(f):
    return len(f)
    
# tests 
print(sommetfile(f1))
print(tailleFile(f1))
```
