### AFVM2 : 
Définir une classe `CompteBancaire()`, qui permette d’instancier des objets tels que `compte1`, `compte2`, etc. 
1. Le constructeur de cette classe initialisera deux attributs d’instance `nom` et `solde`. Trois autres méthodes seront définies :
2. `__str__`		doit renvoyer le nom du titulaire et le solde de son compte.
3. `depot(somme)`	permettra d’ajouter une certaine somme au solde.
4. `retrait(somme)`	permettra de retirer une certaine somme du solde.

On complètera le programme ci-dessous au fur et à mesure que l'on pourra tester sur les exemples proposés plus bas.

```python
class CompteBancaire:
    def __init__(self, nom,solde=0) :
        self.nom = nom
        self.solde = solde
        
    def __str__(self) :
        return "Compte de "+ self.nom + " avec un solde de "+ str(self.solde)+"."
    
    def depot(self,somme):
        self.solde = self.solde + somme
    
    def retrait(self,somme):
        self.solde = self.solde - somme


compte1=CompteBancaire('Prunelle',800)
print(compte1)
compte1.depot(350)
compte1.retrait(20)
print(compte1)

compte2=CompteBancaire("GG")
print(compte2)
compte2.depot(100)
compte2.retrait(150)
print(compte2)
```

### AFVM3:

On modélise un appartement et ses différentes pièces à l’aide des classes `Piece` et `Appartement`, dont on
vous donne les en-têtes de méthodes ci-dessous. L’appartement contient donc un ensemble (de type `list`)
d’objets de type `Piece`.

1. Compléter le code ci dessous

```python
class Piece:
    # nom est un string et surface est un float
    def __init__(self,nom,surface):
        self.nom = nom
        self.surface = surface
    # Accesseurs: retournent les attributs d'un objet de cette classe
    def getSurface(self):
        return self.surface
    def getNom(self):
        return self.nom
    # Mutateur
    def setSurface(self,s): # s est un float,
        if s > 0:
            self.surface = s
            return True
        else :
            return False
            
        
class Appartement:
    # nom est un string, listeDePieces est de type list
    def __init__(self,nom):
        self.listeDePieces=[]# Liste de pièces (objets issus de la classe Piece)
        self.nom=nom
    
    # Accesseurs:
    def getNom(self):
        return self.nom
    
    # Mutateurs
    # pour ajouter une pièce de classe Piece
    def ajouter(self,piece):
        self.listeDePieces.append(piece)
    # pour avoir le nombre de pièces de l'appartement
    def nbPieces(self): #
        return len(self.listeDePieces)
    # retourne la surface totale de l'appartement (un float)
    def surfaceTotale(self):
        surface = 0
        for piece in self.listeDePieces:
            surface += piece.surface
        return surface
            
    # retourne la liste des pièces avec les surfaces
    def getListePieces(self): # sous forme d'une liste de tuples
        liste = []
        for piece in self.listeDePieces:
            liste.append( (piece.nom,piece.surface) )
        return liste
```

    * créer une pièce `cuisine`, de surface 12 m² ;
    * créer un appartement `appart205` qui contiendra toutes les pièces créées ;
    * afficher la surface totale de l'appartement créé ;
    * afficher la liste des pièces et surfaces de l'appartement créé.

```python
piece1 = Piece("chambre1",20)
piece2 = Piece("chambre2",15)
piece3 = Piece("sejour",25)
piece4 = Piece("sdb",10)
piece5 = Piece("cuisine",12)
appart = Appartement("appart205")
appart.ajouter(piece1)
appart.ajouter(piece2)
appart.ajouter(piece3)
appart.ajouter(piece4)
appart.ajouter(piece5)
print(appart.getListePieces())
print(appart.surfaceTotale())
print(appart.nbPieces())
```

4. Faire une fonction `positive(p)` qui crée une pile avec les élements positifs de la pile `p`, dans le même ordre que `p`.

```python
import random as r
def positive(p):
    p1 = Pile()
    p_pos = Pile()
    while not p.estVide():
        s = p.depiler()
        p1.empiler(s)
    while not p1.estVide():
        s = p1.depiler()
        p.empiler(s)
        if s >= 0:
            p_pos.empiler(s)
    return p_pos

pile = Pile()
for i in range(5):
    pile.empiler(r.randint(-10,10))
print(pile)
p_pos = positive(pile)
print(p_pos)
print(pile)
```
