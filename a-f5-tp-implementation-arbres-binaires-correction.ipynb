## I. Implémentation à l'aide du type `list`

L’idée est de représenter un arbre avec un tableau contenant des tableaux.
Un tableau contient trois cases : 
* indice 0 : l’étiquette de la racine ; 
* indice 1 : le sous arbre gauche ; 
* indice 2 : le sous-arbre droit.

Un arbre vide est représenté par un tableau vide.

Ainsi l’arbre ci-dessous est implémenté par la variable `ab` telle que

 `ab = ["A",["B",["D",[],[]],["E",["G",[],[]],[]]],["C",[],["F",[],[]]]]`
 
 
 
Implémenter et tester les fonctions `taille(arbre)` et `hauteur(arbre)` qui renvoient respectivement la taille et la hauteur de l’arbre passé en paramètre.

```python
ab = ["A",["B",["D",[],[]],["E",["G",[],[]],[]]],["C",[],["F",[],[]]]]
```

```python
def taille(arbre): 
    if arbre == [] :
        return 0
    return 1 + taille(arbre[1]) + taille(arbre[2])

def hauteur(arbre): 
    if arbre == [] :
        return -1
    return 1 + max(hauteur(arbre[1]),hauteur(arbre[2]))

print("taille : ",taille(ab))
print("hauteur : ",hauteur(ab))
```

## II. Implémentation en POO

Un arbre binaire vide est implémenté par `None`.

Pour représenter un arbre binaire non vide, nous allons créer une classe `ArbreBinaire` ayant 3 attributs :
* la valeur de la racine de l’arbre binaire ;
* son sous-arbre gauche (`None` par défaut), qui sera une instance de la classe `ArbreBinaire` ;
* son sous-arbre droit (`None` par défaut), qui sera une instance de la classe `ArbreBinaire`.

En plus du constructeur de la classe, on rajoute la méthode ` __str__ ` permettant d’afficher la valeur de la racine de l’arbre comme une chaîne de caractère. Elle est appelée lors de l’utilisation de de la fonction `print` pour afficher un arbre.



```python
class ArbreBinaire :
    def __init__(self, r, g = None, d = None):
        """constructeur"""
        self.racine = r 
        self.gauche = g 
        self.droit = d
    def __str__(self):
        """ renvoie une représentation de l’arbre binaire sous forme de chaine de caracteres de la forme
        < racine , gauche , droit > """
        return '<' + str(self.racine)+','+str(self.gauche)+','+str(self.droit)+'>'
```

5. Redéfinir la méthode `__eq__(self,arbre)` de la classe `ArbreBinaire` de telle sorte que l'appel `A == B` renvoie `True` si les arbres `A` et `B` sont identiques et `False` sinon.

```python
class ArbreBinaire :
    def __init__(self, r, g = None, d = None):
        """constructeur"""
        self.racine = r 
        self.gauche = g 
        self.droit = d
    def __str__(self):
        """ renvoie une représentation de l’arbre binaire sous forme de chaine de caracteres de la forme
        < racine , gauche , droit > """
        return '<' + str(self.racine)+','+str(self.gauche)+','+str(self.droit)+'>'
    def est_une_feuille(self):
        return  self.droit == None and self.gauche == None


    def __eq__(self,arbre):
        if arbre == None :
            return False
        else:
            return self.racine == arbre.racine and self.gauche == arbre.gauche and self.droit == arbre.droit
```
