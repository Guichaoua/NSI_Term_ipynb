## A. `Cellule`
Compléter le code suivant et le tester.

```python
class Cellule :
    def __init__ (self , valeur , suivant ):
        self.v = valeur
        self.s = suivant

    def longueur (self):
        """ renvoie la longueur de la liste dont la cellule
        self est en tête """
        if self.s == None:
            return 1
        else:
            return 1+self.s.longueur()
 
    def ieme (self ,i):
        """ renvoie le i-ième élément de la liste dont la cellule
        self est en tête. Le premier élément est de rang 0 """
        assert i >= 0 , "indice négatif"
        if i==0:
            return self.v
        else:
            return self.s.ieme(i-1)
 
    def affiche ( self ):
        """ renvoie une chaine de caractères avec les valeurs de la liste dont la cellule
        self est en tête. Les valeurs sont représentées dans
        l'ordre séparées par des virgules """
        if self.s == None:
            return  str(self.v)
        else:
            return  str(self.v)+ "," + self.s.affiche()

# on testera le code suivant
l = Cellule(1,Cellule(2,Cellule(3,None)))
l1 = Cellule(3,None)
print(l1.longueur())
print(l.affiche())
print(l.ieme(1))
```

## B. `Liste`

Compléter le code suivant et le tester.

```python
class Liste :
    def __init__ ( self ):
        """ construit une liste vide """
        self.tete = None

    def est_vide ( self ):
        """ renvoie True si la liste est vide et False sinon """
        return self.tete == None

    def ajoute (self ,v):
        """ ajoute une cellule en tête de liste
        avec une valeur v """
        t = Cellule(v,self.tete)
        self.tete = t
 
    def __len__ ( self ):
        """ renvoie la longueur de la liste """
        if self.est_vide ():
            return 0
        else :
            return self.tete.longueur()
 
    def __getitem__ (self ,i):
        """ renvoie le i-ième élément de la liste """
        assert not self.est_vide(), "liste vide"
        return self.tete.ieme(i)
 
    def __str__ ( self ):
        """ affiche les éléments de la liste entre crochets ,
        séparés par des virgules """
        if self.est_vide () :
             return "[]"
        else :
              return self.tete.affiche()
# on testera avec le code suivant
l = Liste()
print(l)
print(l.est_vide())
l.ajoute(3)
print(l[0])
l.ajoute(2)
l.ajoute(1)
print(l)
print(l[0], l[1], l[2])
```

## AFVM1 : Renverser une liste
1. Compléter la fonction `renverser(c)`, pour `c` un objet de type `Cellule`, qui renvoie un nouvel objet
de type `Cellule` avec les éléments de `self` renversés. 

Par exemple, si
`cel = Cellule(1,Cellule(2,Cellule(3,None)))`,
alors `renverser(cel)` renvoie : `Cellule(3,Cellule(2,Cellule(1,None)))`

```python
def renverser(c):
    res = None
    while c != None:
        res = Cellule(c.v,res)
        c = c.s        
    return res

# code à tester
cel = Cellule(1,Cellule(2,Cellule(3,None))) 
renverser(cel).affiche()
        
```

## AFVM3 : Concaténation de listes
1. compléter la fonction récursive `concat(cel1,cel2)`, pour 2 instances de la classe `Cellule`, qui renvoie un nouvel
objet de type `Cellule` avec les éléments de `cel1` et de `cel2` collés bout à bout dans
un nouvel objet de type `Cellule`. Par exemple, si
`cel1 = Cellule(1,Cellule(2,None))`,
et `cel2 = Cellule(3,Cellule(4,None))`, alors `concat(cel1,cel2)` renvoie :
`Cellule(1,Cellule(2,Cellule(3,Cellule(4,None))))`

```python
def concat(cel1, cel2):
    """qui renvoie un nouvel objet de type Cellule 
    avec les éléments de self et de cellule collés bout à bout dans un nouvel objet de type Cellule"""
    if cel1.s == None:
        return Cellule(cel1.v,cel2)
    else:
        return Cellule(cel1.v,concat(cel1.s,cel2))
        

# on testera le code suivant
cel1 = Cellule(1,Cellule(2,Cellule(3,None)))
cel2 = Cellule(3,None)
concat(cel1,cel2).affiche()
```

```python
def trouve(cellule,e):
    if cellule.v==e:
        return True
    elif cellule.s == None:
            return False
    else:
        return trouve(cellule.s,e)
# on testera le code suivant
cel1 = Cellule(1,Cellule(2,Cellule(3,None)))
print(trouve(cel1,2))
print(trouve(cel1,4))
```
