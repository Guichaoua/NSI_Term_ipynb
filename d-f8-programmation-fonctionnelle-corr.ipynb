2. Donner une version itérative de `fibo_rec`.

```python
def fibo_iter(n):
    """TODO Suite de Fibonacci, version itérative"""
    f0 = 0
    f1 = 1
    f2 = 1
    if n <= 1:
            return n
    else:
        for i in range(n-2):
            f0, f1, f2 = f1, f2, f1+f2
    return f2
    
fibo_iter(5)       
```

# AFVM3 : 

On donne la fonction factorielle suivante :


    def fact_iter(n):
        """Factorielle, version itérative"""
        r = 1
        for i in range(1, n + 1):
            r = r*i
        return r

Donner une version recursive de `fact_iter`.

```python
def fact_rec(n):
    """Factorielle, version recursive"""
    if n == 0:
        return 1
    else : 
        return n*fact_rec(n-1)
    
assert fact_rec(3)==3*2*1
assert fact_rec(4)==4*3*2*1
```

2. Écrire une fonction nommée `calc` qui calcule le résultat de $f(x_1) ⊕ f(x_2) ⊕ ... ⊕ f(x_n)$
sur les éléments $x_1$,...,$x_n$ d’un tableau, pour une fonction $f$ et une opération $⊕$ quelconques.

```python
def calc(f,op,t): 
    v = f(t[0])
    for i in range(1,len(t)): 
        v = op(v,f(t[i]))
    return v

def calc2(f,op,t):
    return calcul(op, list(map(f,t)))
```

2. Définir `forall` (version boucle `for`) : 

```python
def forall_for(pred, l):
    """TODO : une version de forall avec une boucle for """
    resultat = True
    for x in l :
        if not pred(x):
            resultat = False
    return resultat

ex1 = [4, 4, 2, 0, 1]
ex2 = []
ex3 = [4, 4, 2, 0, 8]

def pair(x):
    """Un test de parité"""
    return x % 2 == 0
        
print(forall_for(pair, ex1))
print(forall_for(pair, ex2))
print(forall_for(pair, ex3))
```

3. Définir `forall` (version récursive avec les listes). 
Pour cela, on utilisera les fonctions suivantes : 

```python
def empty(l):
    """Retourne True ssi la liste est vide"""
    return len(l) == 0


def head(l):
    """Retourne la tête de la liste, son 1er élément, et None si la liste est vide"""
    if empty(l):
        return None
    return l[0]


def tail(l):
    """Retourne la liste privée de son 1er élément, et None si la liste est vide"""
    if empty(l):
        return None
    return l[1:]


def forall_funct(pred, l):
    """TODO : une version récursive de forall, plus fonctionnelle, utilisant UNIQUEMENT
    empty, head et tail, sans boucle for ni affectation. La fonction sera récursive"""
    if empty(l):
        return True
    else : 
        return pred(head(l)) and forall_funct(pred, tail(l))

ex1 = [4, 4, 2, 0, 1]
ex2 = []
ex3 = [4, 4, 2, 0, 8]

def pair(x):
    """Un test de parité"""
    return x % 2 == 0
        
print(forall_funct(pair, ex1))
print(forall_funct(pair, ex2))
print(forall_funct(pair, ex3))
```

4. Définir `forall` (version avec `map`)

```python
def forall_filter(pred, l):
    """TODO : une version de forall utilisant les fonctions standards filter et len. On pourra avoir
    besoin de transformer le résultat intermédiaire avec list()"""
    return len(list(filter(pred, l)))==len(l) 

ex1 = [4, 4, 2, 0, 1]
ex2 = []
ex3 = [4, 4, 2, 0, 8]

def pair(x):
    """Un test de parité"""
    return x % 2 == 0
        
print(forall_filter(pair, ex1))
print(forall_filter(pair, ex2))
print(forall_filter(pair, ex3))
```

6. Définir `forall` (version avec compréhension de listes)

```python
def forall_comprehension(pred, l):
    """TODO : une version de forall utilisant la compréhension de liste au lieu de map ou filter"""
    return len([elt for elt in l if pred(elt)])==len(l) 

print(forall_comprehension(pair, ex1))
print(forall_comprehension(pair, ex2))
print(forall_comprehension(pair, ex3))
```

7. Définir `forall`  (version avec `reduce`)

La fonction standard `reduce` du module `functools` (appelée aussi `fold`) prend trois paramètres
* une fonction `f`  de type A * B -> A
* une liste `l` de type [A]
* une valeur initiale `x0` de type A

Pour `l = [x1, x2, ..., xn]`, `reduce(f, l, x0)` calcule`f(...(f(f(x0, x1), x2)...), xn)`

Par exemple, `reduce(lambda acc, x : acc * x, range(1,7), 1)` calcule `((((((1*1)*2)*3)*4)*5)*6) = 6! = 120`

https://docs.python.org/3/library/functools.html#functools.reduce

```python
import functools

def forall_reduce(pred, l):
    """TODO : une version de forall utilisant UNIQUEMENT la fonction standard functools.reduce"""
    return functools.reduce(lambda x,y : x and pred(y),l,True)

print(forall_reduce(pair, ex1))
print(forall_reduce(pair, ex2))
print(forall_reduce(pair, ex3))
```
